```python
import matplotlib.pyplot as plt

# Loop over dataframes
for idx, df in enumerate(dataframes, start=1):
    print(f"Processing DataFrame {idx}")

    # Extract features and target variable
    X = df[variable_names].values
    y = df['Y'].values

    # Standardize features and target variable
    scaler_x = StandardScaler()
    scaler_y = StandardScaler()
    X_std = scaler_x.fit_transform(X)
    y_std = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()

    # Build PLS model with the best hyperparameters (use previously tuned parameters)
    pls_model = PLSRegression(n_components=best_n_components, scale=True, tol=best_tol, max_iter=best_max_iter)

    # Fit the PLS model
    pls_model.fit(X_std, y_std)

    # Make predictions with the fitted model
    y_pred_std = pls_model.predict(X_std)

    # Inverse transform predictions back to the original scale
    y_pred = scaler_y.inverse_transform(y_pred_std.reshape(-1, 1)).flatten()
    y_actual = scaler_y.inverse_transform(y_std.reshape(-1, 1)).flatten()

    # Plot actual vs predicted values for the dataset
    plt.figure(figsize=(8, 6))
    plt.scatter(y_actual, y_pred, color='b', alpha=0.7, label='Predicted vs Actual')
    plt.plot([min(y_actual), max(y_actual)], [min(y_actual), max(y_actual)], color='r', linestyle='--', label='Perfect Prediction')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title(f'Prediction Results - Dataset {idx}')
    plt.legend()
    plt.grid(True)

    # Add RMSE annotation to the plot
    rmse_value = np.sqrt(np.mean((y_actual - y_pred) ** 2))  # Calculate RMSE
    plt.text(0.05, 0.9, f'RMSE: {rmse_value:.4f}', 
             fontsize=10, ha='left', va='top', transform=plt.gca().transAxes, 
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1', alpha=0.5))

    # Show the plot for this dataset
    plt.tight_layout()
    plt.show()
```
